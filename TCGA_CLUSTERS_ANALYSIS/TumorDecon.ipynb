## Now, we will analyze the results and graphically represent them

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
import numpy as np
import glob
```

```python
# Function that will be used to calculate the adjusted p values by Benjamini-Hochberg
def fdr(p_vals):

    from scipy.stats import rankdata
    ranked_p_values = rankdata(p_vals)
    fdr = p_vals * len(p_vals) / ranked_p_values
    fdr[fdr > 1] = 1

    return fdr
```

```python
# Define a function that charges the datafame and includes a cluster column
def charge_df_results(path,cluster_dict=None):

    df = pd.read_csv(path,sep="\t")
    df.Patient_ID = df.Patient_ID.str.replace("01A","01")
    if cluster_dict!=None:
        # Add the cluster anotation
        df["Cluster"]= df.Patient_ID.map(cluster_dict)
    return df
```

```python
def boxplots_td(df_boxplot,ylabel,title):
    # Create and save the figure
    plt.figure(figsize=(25,8))
    ax = sns.boxplot(data=df_boxplot, x="Cell Type", y="Frequency", hue="Group")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right') # rotate the x labels
    plt.title(f"Deconvolution by {title}", fontsize=20,fontweight=800)
    plt.xlabel("CELL TYPE",fontsize = 15) 
    plt.ylabel(ylabel,fontsize = 15)
    plt.savefig(f"./Results/TumorDecon/Tumordecon_{title}_boxplot.png",bbox_inches="tight")
    plt.show()
```

```python
# List of result file paths
plot_files = glob.glob("./Results/*results.csv")
```

```python
# Charge the group labels
df_cls = pd.read_csv("./Data/Common_data/Clusters.csv",sep=",").astype(str)

# Change the group labels
df_cls.Group = df_cls.Group.str.replace("1","Cluster 1") 
df_cls.Group = df_cls.Group.str.replace("0","Cluster 0")

# Create a dictionary to map this labels into the results datafames to plot
dict_cls= dict(zip(df_cls["Unnamed: 0"],df_cls.Group))
```

## Calculate the stastics

Now we evaluate if there are signifficant differences in the scoring between clusters and within cell types and create a report for every result file.

```python

for n,plot_file in enumerate(plot_files):
    # Charge the tumor deconv data
    df = charge_df_results(plot_files[0],dict_cls)

    if n == 0:
        # Define a dictionary that will contain the statistics resuts 
        ttest_dict = {"Cell Type": df.columns[1:-1].values}

    # Initialize the lists that will contain the plot, and stastics values
    class_list,counts_list,group_list,pvalues,difference,shapiro = [],[],[],[],[],[]
    
    for i in df.columns[1:-1]:

        cell_data = df[i]
        groups_data = df["Cluster"]

        # The boxplot colums creation
        class_list= class_list + [i]*len(df)
        counts_list = counts_list + cell_data.tolist()
        group_list = group_list + groups_data.tolist()

        # TESTING THE NORMALITY
        data_0 = cell_data[groups_data=="Cluster 0"].values
        data_1 = cell_data[groups_data=="Cluster 1"].values

        # Store the pvalue of the normality test for data of both clusters
        shapiro.append([sp.stats.shapiro(data_0).pvalue,sp.stats.shapiro(data_1).pvalue]) 

    # Correct the normality tests by the BH fdr 
    shadj = fdr(np.ravel(shapiro))<0.05

    # This is for the T-test
    for ind ,i in enumerate(df.columns[1:-1]):

        cell_data = df[i]
        groups_data = df["Cluster"]

        # The t test
        data_0 = cell_data[groups_data=="Cluster 0"].values
        data_1 = cell_data[groups_data=="Cluster 1"].values

        # USE TTEST OR MANNWHITNEY TESTS REGARDING THE NORMALITY OF THE DATA

        if not (shadj[ind*2] or shadj[ind*2+1]): # Both data are normal
            pvalues.append(sp.stats.stats.ttest_ind(data_1,data_0).pvalue)

        else: # Data not normally distributed
            pvalues.append(sp.stats.mannwhitneyu(data_1,data_0).pvalue)

        difference.append(np.mean(data_1)-np.mean(data_0))

    # Calculate the BH adjusted pvals
    padj = fdr(np.array(pvalues))

    # Now we save the ttest results in the ttest dict
    title = plot_file.split("_")[0].split("/")[1].upper()

    ttest_dict[f'{title}_mean_diff']  =  difference
    ttest_dict[f'{title}_padj'] = padj
    ttest_dict[f'{title}_significative'] = padj<0.05
```

```python
# Create the stastics dataframe for all the deconv methods and save it
df_ttest = pd.DataFrame(ttest_dict)
df_ttest.to_csv("./Results/TumorDecon/tumordecon_stastics.tsv",sep="\t",index=False)
df_ttest.head()
```

# Filter out the celltypes with no significant or contradictory differences 

```python
# With this function we can extract the cells that change in the same direction and are significative with both methods
def common_patterns(row):
    sig = row["DECON_significative"] & row["SINGSCORE_significative"]
    diff = (0-row["DECON_mean_diff"])*(0-row["SINGSCORE_mean_diff"]) > 0
    return sig & diff

# Apply the function and extract the cells with a common pattern
common_pattern_cells = df_ttest.loc[df_ttest.apply(common_patterns,axis=1)]["Cell Type"].tolist()
```

## Create the boxplots

```python
for n,plot_file in enumerate(plot_files):
    # Charge the tumor deconv data
    df = charge_df_results(plot_files[0],dict_cls)

    # Initialize the lists that will contain the plot, and stastics values
    class_list,counts_list,group_list = [],[],[]

    for i in df.columns[1:-1]:

        cell_data = df[i]
        groups_data = df["Cluster"]

        # The boxplot colums creation
        class_list= class_list + [i]*len(df)
        counts_list = counts_list + cell_data.tolist()
        group_list = group_list + groups_data.tolist()


    # Define the boxplot dataframe 
    df_boxplot= pd.DataFrame({"Cell Type":class_list,"Frequency":counts_list,"Group":group_list})
    
    # Extract the common pattern cells calculated before
    df_boxplot = df_boxplot[df_boxplot["Cell Type"].isin(common_pattern_cells)] 
    
    title = plot_file.split("_")[0].split("/")[1].upper()
    ylabel="FREQUENCY"
    if title=="SINGSCORE": # Singscore provide ssGSEA scores not cell frequencies
        ylabel="SCORE"

    boxplots_td(df_boxplot,ylabel,title) # Create the boxplot 
```
