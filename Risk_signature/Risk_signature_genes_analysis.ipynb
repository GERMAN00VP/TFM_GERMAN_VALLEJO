```python
import pandas as pd
import pickle as pkl 
from utils import Risk_signature
import matplotlib.pyplot as plt
```

```python
with open("./Data/Datasets_dict.pkl","rb") as f:

    datasets = pkl.load(file=f) # Charge the datasets
```

```python
res,more_res= [],[]
Risk_signature.create_results_folder("./Results/Risk_sig_post_analysis") # Create the results folder
genes= pd.read_csv("./Results/Risk_signature_TFM/Multivariant_Cox_significative_results.csv").covariate.tolist() # Charge the genes of the risk signature

for gen in genes:
    
    Risk_signature.create_results_folder(f"./Results/Risk_sig_post_analysis/{gen}")

    # Iterate the different datasets
    for key in datasets.keys():
        df = datasets[key]
        df = df[["OS","OS.time",gen]] # Filter out the genes we are not using

        df[gen].hist(bins=30) # Represent the gene expression distribution
        plt.axvline(df[gen].median(),color="black")
        plt.savefig(f"./Results/Risk_sig_post_analysis/{gen}/Histograme_{gen}_{key}_dataset",bbox_inches="tight")
        plt.close()
        
        # This stratificates the patients in high and low gene expression levels (the median is the threshold) and plots the KM curves
        res.append(Risk_signature.Risk_Sign_Calculator(LASSOcoefs=[1],LASSOgenes=[gen],df=df,
                                            save=f"./Results/Risk_sig_post_analysis/{gen}/KM_plot_{gen}_{key}_dataset",show=False) )
        more_res.append((gen,key)) # Save the genes and datasets tested
        plt.close()
    
```

```python
# Create a dataframe with the results of the analysis and concatenate the gene and datasets info
resdf = pd.DataFrame(res,columns=["Survival High Expression - Survival Low Expression ","Log Rank Test P-value"])
resdf[["Gene","Dataset"]]= pd.DataFrame(more_res,columns=["Gene","Dataset"])

# Rename the datasets
resdf.Dataset.replace("ICGC_ARRAYS","ICGC-AU2",inplace=True)
resdf.Dataset.replace("AU","ICGC-AU",inplace=True)
resdf.Dataset.replace("CA","ICGC-CA",inplace=True)

# Create an easy to read table
resdf = round(resdf.pivot(index="Gene",columns="Dataset",values=["Survival High Expression - Survival Low Expression ","Log Rank Test P-value"]),4)
resdf.to_csv("./Results/Risk_sig_post_analysis/Sumary_results_table.csv")
resdf
```
