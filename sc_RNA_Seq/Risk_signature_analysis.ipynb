```python
import scanpy as sc
```

```python
adata= sc.read_h5ad("./Data/Peng_annotated.h5ad")
```

```python
adata.obs["CONDITION"] = adata.obs.Condition.replace("T","Tumor tissue")
adata.obs["CONDITION"] = adata.obs.CONDITION.replace("N","Control tissue")
adata.obs.CONDITION
```

```python
# List of genes selected from the risk signature
keys = ["LYZ","CONDITION","FGF13"]
```

```python
for key in keys:
    sc.pl.umap(adata,color=key,frameon=False)
```
